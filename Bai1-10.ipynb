```python
#bai9 binary exponentiation (lũy thừa nhị phân )  tối ưu hơn
a,b=map(int,input().split())
def binaryexpo(a,b):
    if b==0:
        return 1
    else:
        x=(a**(b//2))#tìm được biến dùng cho cả 2 hạn chế tính toán 
        if b%2==0:
            return x *x
        else:
            return x*x*a
if __name__=="__main__":
    print(binaryexpo(a,b))
        
```

```python
def generate_next_binary_string(A):
    """
    Sinh chuỗi nhị phân kế tiếp của A theo thứ tự từ điển.
    A là một list (mảng) biểu diễn chuỗi nhị phân.
    Trả về True nếu sinh thành công, False nếu A là chuỗi cuối cùng (toàn 1).
    """
    n = len(A)
    i = n - 1  # Bắt đầu tìm kiếm từ phần tử cuối cùng (index n-1)

    # 1. Tìm vị trí i cuối cùng mà A[i] = 0 (từ phải sang trái)
    while i >= 0 and A[i] == 1:
        i -= 1

    # 2. Kiểm tra điều kiện dừng: Nếu i < 0, chuỗi hiện tại là [1, 1, ..., 1] -> END
    if i < 0:
        return False  # Không có chuỗi kế tiếp

    # 3. Sinh chuỗi kế tiếp:
    # a. Đảo A[i] từ 0 thành 1
    A[i] = 1

    # b. Đặt tất cả các phần tử phía sau A[i] về 0
    # Lưu ý: Do vòng while đã đảm bảo A[i+1] đến A[n-1] đều là 1 (trước khi đổi A[i])
    # nên ta chỉ cần đổi chúng về 0.
    for j in range(i + 1, n):
        A[j] = 0

    return True  # Sinh thành công

def list_all_binary_strings(n):
    """
    Liệt kê tất cả 2^n chuỗi nhị phân độ dài n.
    """
    # Khởi tạo chuỗi đầu tiên: [0, 0, ..., 0]
    A = [0] * n
    count = 0

    while True:
        count += 1
        # In chuỗi hiện tại
        print(f"Chuỗi #{count}: {''.join(map(str, A))}")

        # Cố gắng sinh chuỗi kế tiếp
        if not generate_next_binary_string(A):
            break  # Nếu không sinh được, dừng vòng lặp

# Ví dụ minh họa
N = 3
print(f"--- Liệt kê tất cả chuỗi nhị phân độ dài n = {N} ---")
list_all_binary_strings(N)
```
