```python
#01
n=int(input())
def tinhtongn(a):
    if a==0:
        return 0
    else:
        return a + tinhtongn(a-1)
if __name__=="__main__":
    print(tinhtongn(n))
    
```

```python
n=int(input())
def bai2(a):
    if a==1:
        return 1
    else:
        return a**2+bai2(a-1)
        # ko phải bai2(a-1)**2
if __name__== "__main__":
    print(bai2(n))
    
```

```python
#bai3
n=int(input())
def bai3(a):
    if a==1:
        return 1
    else:
        return a*a*a +bai3(a-1)
if __name__=="__main__":
    print(bai3(n))
```

```python
# bai4
n=int(input())
def bai4(a):
    if a==1:
        return -1
    else:
        if a%2==0:
            return a+ bai4(a-1)
        else:
            return -(a) +bai4(a-1)
if __name__=="__main__":
    print(bai4(n))
    
```

```python
# bai 5 giai thua
n=int(input())
def bai5(a):
    if a==0:
        return 1# bởi vì nếu =0 sẽ tiếp tục lặp từ vòng -1 đến âm vô cùng
    else:
        return a *bai5(a-1)
if __name__=="__main__":
    print(bai5(n))
    
```

```python
#bai6
n=int(input())
def bai6(a):
    if a==0:
        return 0
    if a==1:
        return 1
    else:
        return bai6(a-1)+bai6(a-2)
        
        

if __name__=="__main__":
    print(bai6(n))
```

```python
from math import *
n,k=map(int,input().split())
def bai7(n,k):
    if k==0 or n==k:
        return 1
    else:
        return bai7(n-1,k-1)+bai7(n-1,k)
if __name__=="__main__":
    print(bai7(n,k))
```

```python
#bai8 ước chung lớn nhất và bội chung nhỏ nhất 
a,b=map(int,input().split())
def bai8(a,b):
    if b==0:
        return a # để nó xử lý nếu có số 0 trong cả a và b , ta có thể thay a=b và b =a cũng được     
    else:
        return bai8(b,a%b)#thuật toán ecluid
if __name__=="__main__":
    print(bai8(a,b), a*b// bai8(a,b))
```

```python
#bai9 binary exponentiation (lũy thừa nhị phân ) chưa tối ưu
a,b=map(int,input().split())
def binaryexponentiation(a,b):
    if b==0:
        return 1
    else:
        if b%2==0:
            return (a**(b/2)) *(a **(b/2))#bạn đang bắt máy tính phải tính 2 lần giống nhau ==> tốn time
        else:
            return (a**((b-1)/2)) *(a **((b-1)/2)) *(a)#ta có thể thay phép chia nguyên để sử dụng một biến cùng với nếu b chẵn 
if __name__=="__main__":
    print(binaryexponentiation(a,b))
```

```python
#bai9 binary exponentiation (lũy thừa nhị phân )  tối ưu hơn
a,b=map(int,input().split())
def binaryexpo(a,b):
    if b==0:
        return 1
    else:
        x=(a**(b//2))#tìm được biến dùng cho cả 2 hạn chế tính toán 
        if b%2==0:
            return x *x
        else:
            return x*x*a
if __name__=="__main__":
    print(binaryexpo(a,b))
        
```
