```python
#thuật toán sinh hoán vị kế tiếp 
def next_temptation(P):
    n=len(P)
    k=-1
    for i in range(n-2,-1,-1):
        if P[i]<P[i+1]:
            k=i
            break
    if k==-1:
        return False
    j=-1
    for i in range(n-1,k,-1):
        if P[i]>P[k]:
            j=i
            break
    P[k],P[j]=P[j],P[k]
    P[k+1:]=P[k+1:][::-1]
    return True
def print_put():
    nhap=input()
    P=[]
    for i in nhap:
        P.append(int(i))
    while True:
        print(P)
        if not next_temptation(P):
            break
if __name__=="__main__":
    print_put()
        
            
```
